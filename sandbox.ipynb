```python
#Import trig stuff from math
from math import sin, cos, sqrt, atan2,radians

#Distance function between two lat/lon
def get_dist(lat1, lon1, lat2, lon2):
  r = 6373.0

  lat1 = radians(lat1)
  lon1 = radians(lon1)
  lat2 = radians(lat2)
  lon2 = radians(lon2)

  dlon = lon2 - lon1
  dlat = lat2 - lat1

  a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
  c = 2 * atan2(sqrt(a), sqrt(1 - a))

  return r * c

def filter_by_dist(row, lat2, lon2, max_dist):
    if get_dist(row['latitude'],row['longitude'],lat2,lon2) < max_dist:
        return True
    else:
        return False
```

```python
from dataclasses import dataclass


@dataclass
class Property:
    price: float
    room_type: str
    accommodates: float
    bedrooms: float


@dataclass
class Location:
    latitude: float
    longitude: float
    tolerance: float = 0.0
```

```python
def df_of_properties_in_proximity(df: pd.DataFrame, location: Location, inverse_selection: bool = False) -> pd.DataFrame:
  if inverse_selection:
        filtered_df = df[~df.apply(filter_by_dist, args = (location.latitude, location.longitude, location.tolerance), axis=1)]
  else:
        filtered_df = df[df.apply(filter_by_dist, args = (location.latitude, location.longitude, location.tolerance), axis=1)]
  return filtered_df
```

```python
def df_matching_properties(df: pd.DataFrame, property: Property, price_tolerance: float) -> pd.DataFrame:
  filtered_df = df[(df.price <= property.price*(1.0 + price_tolerance)) & (df.price >= property.price*(1.0 - price_tolerance)) & (df.room_type == property.room_type) & (df.accommodates == property.accommodates) & (df.bedrooms == property.bedrooms)]
  return filtered_df
```

```python
def score_for_coordinates(location: Location, df: pd.DataFrame) -> float:
    num_properties_in_proximity = len(df_of_properties_in_proximity(df, location))
    num_properties_outside_proximity = len(df_of_properties_in_proximity(df, location, inverse_selection=True))
    return num_properties_in_proximity/(num_properties_in_proximity + num_properties_outside_proximity)
```

```python
from typing import Dict


def get_most_likely_location(df: pd.DataFrame, property: Property):
    df_filtered_by_properties = df_matching_properties(df, property, price_tolerance=0.1)

    max_lon = df_filtered_by_properties.longitude.max()
    min_lon = df_filtered_by_properties.longitude.min()
    diff_lon = max_lon - min_lon
    max_lat = df_filtered_by_properties.latitude.max()
    min_lat = df_filtered_by_properties.latitude.min()
    diff_lat = max_lat - min_lat
    steps = 20

    coordinates_to_scan = {}
    for step_lon in range(0, steps+1):
        this_lon = min_lon + step_lon * diff_lon/steps
        for step_lat in range(0, steps+1):
            this_lat = min_lat + step_lat * diff_lat/steps
            this_loc = Location(longitude=this_lon, latitude=this_lat, tolerance=1.0)
            coordinates_to_scan[(this_loc.latitude, this_loc.longitude)] = score_for_coordinates(this_loc, df_filtered_by_properties)
    max_key = max(coordinates_to_scan, key=coordinates_to_scan.get)
    return max_key, coordinates_to_scan[max_key], coordinates_to_scan
```

```python
given_price = 136
given_room_type = "Entire home/apt"
given_accommodates = 2.0
given_bedrooms = 1.0
example_property = Property(price=given_price, room_type=given_room_type, accommodates=given_accommodates, bedrooms=given_bedrooms)
```

```python
coordinates, probability, all_coordinates = get_most_likely_location(df, example_property)
```

```python
this_list = [[coordinates[0], coordinates[1], value] for coordinates, value in all_coordinates.items()]
this_list
```

```python
result_df = pd.DataFrame(this_list, columns=['latitude', 'longitude', "probability"])
result_df
```
